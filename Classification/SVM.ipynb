```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from mlxtend.plotting import  plot_decision_regions
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
X, y = make_classification(n_samples= 100, n_features=   2, n_redundant= 0, n_clusters_per_class=1)
X_train, X_test, y_train , y_test = train_test_split(X, y,test_size= 0.3, random_state= 42, stratify= y)
```

```python
svm = SVC(kernel= 'rbf',C =1.0, gamma= 'scale')
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
plot_decision_regions(X, y, clf = svm, legend= 2)
```

```python
DecisionBoundaryDisplay.from_estimator(
    svm, X, response_method='predict',
    alpha = 0.3, grid_resolution= 100
) ## shows the region with decision boundary not the data points
plt.scatter(X[:, 0], X[:, 1], edgecolors= 'k', cmap= 'coolwarm')
plt.title("SVM decision Boundary")
plt.show()
```
