## fit vs fit_transform

```python
## fit --- just for learning
## transform ---applying what has learned (used it test_set)
## fit_transform -- Learning + applying (used in train_set)
# fit = study
# transform = use what was studied
# fit_transform = study + use immediately
```

```python
## say for example
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X, y = load_iris(return_X_y= True)
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state= 42, stratify= y)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# ## Transformers

# Transformer	What fit() Learns
# StandardScaler:	Mean and standard deviation
# MinMaxScaler:Min and max values
# PCA	 : Principal components
# OneHotEncoder:	Categories
# LabelEncoder:	Unique labels
```
