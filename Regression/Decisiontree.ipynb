```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

# Create sample regression data
X, y = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 42)

# Create and train regression tree
regressor = DecisionTreeRegressor(max_depth=3, random_state=42)
regressor_2 = DecisionTreeRegressor(max_depth=5, random_state= 42)
regressor_2.fit(X_train, y_train)
regressor.fit(X_train, y_train)
```

```python
y1_pred = regressor.predict(X_test)
y2_pred = regressor_2.predict(X_test)
```

```python
plt.figure()
plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data")
plt.plot(X_test, y1_pred, color="cornflowerblue", label="max_depth=2", linewidth=2)
plt.plot(X_test, y2_pred, color="yellowgreen", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import make_regression

# Your exact dataset
X, y = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)

# Your exact regressors
regressor   = DecisionTreeRegressor(max_depth=2, random_state=42)
regressor_2 = DecisionTreeRegressor(max_depth=5, random_state=42)
regressor_2.fit(X, y)
regressor.fit(X, y)

# Fine grid for the step-function lines
X_test = np.linspace(X.min(), X.max(), 500).reshape(-1, 1)

fig, ax = plt.subplots(figsize=(7, 5))
ax.scatter(X, y, s=20, edgecolor='black', facecolor='#c8622a', label='data', zorder=3)
ax.plot(X_test, regressor.predict(X_test),   color='steelblue',   linewidth=2, label='max_depth=2')
ax.plot(X_test, regressor_2.predict(X_test), color='yellowgreen', linewidth=2, label='max_depth=5')

ax.set_title('Decision Tree Regression')
ax.set_xlabel('data')
ax.set_ylabel('target')
ax.legend(framealpha=0.9)
plt.tight_layout()
plt.show()
```
