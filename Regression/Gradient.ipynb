```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_regression
##make_regression is used to generate random regression data set to train or learn regression models
X,y = make_regression(
    n_samples=100,
    n_features=1,
    n_informative= 1,
    random_state= 42,
    noise= 30
)
plt.scatter(X, y)
plt.show()
```
