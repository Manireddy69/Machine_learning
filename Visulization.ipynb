```python
import numpy as np
%matplotlib inline
import matplotlib.pylab as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,3,4,5,6])
plt.plot(x,y) ## plot(x, y)
plt.show()
```

```python
np.arange(1, 10,2) #(start, stop, step)
np.linspace(1,10, 10) #(start, stop, num)
```

```python
x = np.linspace(1, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.show()
```

```python
plt.figure()
x = np.linspace(0,10, 100)
plt.subplot(2,1, 1) #(rows, cols, panelnumber)
plt.plot(x, np.sin(x), '-')
plt.title('Sin x')

plt.subplot(2,1,2)
plt.plot(x, np.cos(x), '--')
plt.title('Cos x')
plt.show()
```

```python
## object-oriented interface
x = np.linspace(0, 2 * np.pi, 360)
fig , ax = plt.subplots(2) ## 2 subplots
ax[0].plot(x, np.sin(x))
ax[0].set_title('Sin')
ax[1].plot(x, np.cos(x))
ax[1].set_title('Cos')

plt.suptitle("Trignometric function", fontsize = 16, fontweight = 'bold')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('ggplot')
plt.plot(x, np.sin(x), '-g', label = 'sin(x)')
plt.plot(x, np.cos(x), '-r', label = 'Cos(x)')
plt.axis('equal') ## x & y axis are equal
plt.legend() ## when we have mutiple data series
plt.show()
```

```python
# • plt.xlabel() → ax.set_xlabel()
# • plt.ylabel() → ax.set_ylabel()
# • plt.xlim() → ax.set_xlim()
# • plt.ylim() → ax.set_ylim()
# • plt.title() → ax.set_title()
```

```python
## histograms, binnings, density

data = np.random.randn(1000)
plt.hist(data, color= 'm')
plt.show()
```

```python
x1= np.random.normal(0,0.5,1000)
x2 = np.random.normal(0,1, 1000)

plt.hist(x1, alpha = 0.5,label= 'x1', color= 'blue')
plt.hist(x2, alpha = 0.5,label= 'x2', color= 'orange') ## alpha removes the transparency
plt.show()
```
