## 📊 Statistical Methods to Detect Outliers


### 🔹 1. Z-Score Method

```python
from scipy import stats

z_scores = np.abs(stats.zscore(df['GrLivArea']))
df_z = df[(z_scores < 3)]  # Keep only data within 3 standard deviations
```

🔍 A Z-score > 3 is considered an outlier for normally distributed data.

### 🔹 2. IQR (Interquartile Range) Method

```python
Q1 = df['GrLivArea'].quantile(0.25)
Q3 = df['GrLivArea'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_iqr = df[(df['GrLivArea'] >= lower_bound) & (df['GrLivArea'] <= upper_bound)]
```

📌 IQR is useful for skewed distributions and is more robust than the Z-score method.

### ✂️ 3. Winsorization (Capping Outliers)

```python
from scipy.stats.mstats import winsorize

# Capping the extreme 5% on both ends
df['GrLivArea_winsor'] = winsorize(df['GrLivArea'], limits=[0.05, 0.05])
```

🛡 Winsorization replaces extreme values with nearest acceptable values — preserving dataset size while reducing outlier impact.

```python
# Before
plt.subplot(1, 2, 1)
sns.histplot(df['GrLivArea'], kde=True, color='orange')
plt.title("Before Handling Outliers")


# After (IQR or Winsorization)
plt.subplot(1, 2, 2)
sns.histplot(df_iqr['GrLivArea'], kde=True, color='seagreen')
plt.title("After Handling Outliers")
plt.show()
```

<div style="background-color:#eafaf1; padding: 15px; border-left: 5px solid #27ae60; border-radius: 8px">
✅ Summary <br>
Use boxplots for quick visual detection.<br>
Use Z-score for normally distributed data.<br>
Use IQR for skewed distributions.<br>
Use winsorization/capping when you need to keep all rows but limit extreme values.

## import necassary libraries

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
import shap
import joblib
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
## XGBoost accepts only int, float, bool, or category dtypes

# Identify all object and category columns
object_cols = X_train.select_dtypes(include=['object', 'datetime64']).columns

# Convert them to category dtype
for col in object_cols:
    X_train[col] = X_train[col].astype('category')
    X_test[col] = X_test[col].astype('category')
```

```python
house_model = xgb.XGBRegressor(
    n_estimators = 100, 
    learning_rate = 0.1,
    max_depth = 6,
    random_state = 42,
    enable_categorical = True
)
house_model.fit(X_train, y_train)
```

## Predict the model

```python
y_train_pred = house_model.predict(X_train)
y_test_pred = house_model.predict(X_test)

## evaluate the model

train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
train_mae = mean_absolute_error(y_train, y_train_pred)
test_mae = mean_absolute_error(y_test, y_test_pred)
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
```

## Visulaization

```python
from sklearn.model_selection import GridSearchCV

parameters = {
    'max_depth': [4, 6, 8],
    'learning_rate': [0.05, 0.1, 0.2],
    'n_estimators': [100, 200],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

grid_search = GridSearchCV(
    house_model,
    parameters,
    cv = 5,
    scoring = 'r2',
    verbose = 1,
    n_jobs = 1
)
grid_search.fit(X_train, y_train)
```

```python
## save the model
best_model = grid_search.best_estimator_

joblib.dump(best_model, "best_house_model.joblib")
```

```python
# Test the best model
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
test_r2 = r2_score(y_test, y_pred)
test_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
# Convert CV results to DataFrame
results_df = pd.DataFrame(grid_search.cv_results_)

# Plot parameter impact
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Effect of max_depth
depth_effect = results_df.groupby('param_max_depth')['mean_test_score'].mean()
axes[0, 0].plot(depth_effect.index, depth_effect.values, 'o-', linewidth=2, markersize=8)
axes[0, 0].set_xlabel('max_depth', fontsize=11)
axes[0, 0].set_ylabel('Mean CV R²', fontsize=11)
axes[0, 0].set_title('Effect of Tree Depth', fontsize=12)
axes[0, 0].grid(True, alpha=0.3)

# Effect of learning_rate
lr_effect = results_df.groupby('param_learning_rate')['mean_test_score'].mean()
axes[0, 1].plot(lr_effect.index, lr_effect.values, 'o-', linewidth=2, markersize=8)
axes[0, 1].set_xlabel('learning_rate', fontsize=11)
axes[0, 1].set_ylabel('Mean CV R²', fontsize=11)
axes[0, 1].set_title('Effect of Learning Rate', fontsize=12)
axes[0, 1].grid(True, alpha=0.3)

# Effect of n_estimators
est_effect = results_df.groupby('param_n_estimators')['mean_test_score'].mean()
axes[1, 0].bar(est_effect.index, est_effect.values, width=30)
axes[1, 0].set_xlabel('n_estimators', fontsize=11)
axes[1, 0].set_ylabel('Mean CV R²', fontsize=11)
axes[1, 0].set_title('Effect of Number of Trees', fontsize=12)
axes[1, 0].grid(True, alpha=0.3)

# Best parameters text
best_params_text = "\n".join([f"{k}: {v}" for k, v in grid_search.best_params_.items()])
axes[1, 1].text(0.1, 0.5, f"BEST PARAMETERS:\n\n{best_params_text}", 
                fontsize=11, verticalalignment='center')
axes[1, 1].axis('off')
axes[1, 1].set_title('Optimal Configuration', fontsize=12)

plt.tight_layout()
plt.show()

print(f"\n✨ Best CV R²: {grid_search.best_score_:.4f}")
print(f"🔧 Best parameters: {grid_search.best_params_}")
```
